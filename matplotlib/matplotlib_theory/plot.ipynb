# Matplotlib tutorial

```python
import matplotlib.pyplot as plt
import numpy as np

# create figure and axes
fig, ax = plt.subplots()

# created 1 figure and 4 axes
fig, ax = plt.subplots(2, 2)


```

```python
# separately
figure = plt.figure()
ax1 = figure.add_subplot(2, 2, 1)
ax2 = figure.add_subplot(2, 2, 2)
ax2 = figure.add_subplot(2, 2, 3)
```

```python
# use coords
fig, ax = plt.subplots()
x = [-3, -2, -1, 0, 1, 2, 3]
y = [9, 4, 1, 0, 1, 4, 9]

ax.plot(x, y) # == plt.plot(x,y) if only 1 plot
```

```python
fig, ax = plt.subplots()

x = [-3, -2, -1, 0, 1, 2, 3]
y = [9, 4, 1, 0, 1, 4, 9]
x_abs = [3, 2, 1, 0, 1, 2, 3]

ax.plot(x, y)
ax.plot(x, x_abs)

```

```python
# using plot matrix

fig, axes = plt.subplots(2, 2)

x = [-3, -2, -1, 0, 1, 2, 3]
y = [9, 4, 1, 0, 1, 4, 9]

axes[1, 0].plot(x, y)
```

```python
# plot stiles

fig, axes = plt.subplots()

x = np.array([-3, -2, -1, 0.1, 1, 2, 3])
y = np.array([9, 4, 1, 0.1, 1, 4, 9])

axes.plot(x, y)
axes.plot(x, np.sin(x), color=(1.0, 0.2, 0.3), linestyle='-')
```

```python
# With points

fig, ax = plt.subplots()

x = np.array([-3, -2, -1, 0.1, 1, 2, 3])
y = np.array([9, 4, 1, 0.1, 1, 4, 9])

ax.plot(x, y, marker='o')
ax.plot(x*0.5, y, marker='^')
ax.plot(2 * x, y, marker='*')
```

```python
fig, ax = plt.subplots()

x = np.array([-3, -2, -1, 0.1, 1, 2, 3])
y = np.array([9, 4, 1, 0.1, 1, 4, 9])

ax.plot(x, y, color='red', linestyle='--', marker='o', alpha=0.2)
ax.plot(0.5*x, y, '*--g', linewidth=2.5)
```

```python
# axis limits
fig, axes = plt.subplots()

x = np.array([-3, -2, -1, 0.1, 1, 2, 3])
y = np.array([9, 4, 1, 0.1, 1, 4, 9])

axes.plot(x, y, color='red', linestyle='--', marker='o')
axes.set_xlim(0, 5)
axes.set_ylim(2, 10)
```

```python
# To mark our plots

fig, ax = plt.subplots()

x = np.array([-3, -2, -1, 0.1, 1, 2, 3])
y = np.array([9, 4, 1, 0.1, 1, 4, 9])

ax.plot(x, y, color='red', linestyle='-', marker='o', label='y = x^2')
ax.plot(0.5 * x, y, color='green', linestyle='-', marker='*', label='y = 0.5 * x^2')

ax.set_title('Header of our plot', fontsize=20)

ax.set_xlabel("x")
ax.set_ylabel("y")

ax.legend(loc='lower left')
```
