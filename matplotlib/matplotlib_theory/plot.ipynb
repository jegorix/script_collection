# Matplotlib tutorial

```python
import matplotlib.pyplot as plt
import numpy as np

# create figure and axes
fig, ax = plt.subplots()

# created 1 figure and 4 axes
fig, ax = plt.subplots(2, 2)


```

```python
# separately
figure = plt.figure()
ax1 = figure.add_subplot(2, 2, 1)
ax2 = figure.add_subplot(2, 2, 2)
ax2 = figure.add_subplot(2, 2, 3)
```

```python
# use coords
fig, ax = plt.subplots()
x = [-3, -2, -1, 0, 1, 2, 3]
y = [9, 4, 1, 0, 1, 4, 9]

ax.plot(x, y) # == plt.plot(x,y) if only 1 plot
```

```python
fig, ax = plt.subplots()

x = [-3, -2, -1, 0, 1, 2, 3]
y = [9, 4, 1, 0, 1, 4, 9]
x_abs = [3, 2, 1, 0, 1, 2, 3]

ax.plot(x, y)
ax.plot(x, x_abs)

```

```python
# using plot matrix

fig, axes = plt.subplots(2, 2)

x = [-3, -2, -1, 0, 1, 2, 3]
y = [9, 4, 1, 0, 1, 4, 9]

axes[1, 0].plot(x, y)
```

```python
# plot stiles

fig, axes = plt.subplots()

x = np.array([-3, -2, -1, 0.1, 1, 2, 3])
y = np.array([9, 4, 1, 0.1, 1, 4, 9])

axes.plot(x, y)
axes.plot(x, np.sin(x), color=(1.0, 0.2, 0.3), linestyle='-')
```

```python
# With points

fig, ax = plt.subplots()

x = np.array([-3, -2, -1, 0.1, 1, 2, 3])
y = np.array([9, 4, 1, 0.1, 1, 4, 9])

ax.plot(x, y, marker='o')
ax.plot(x*0.5, y, marker='^')
ax.plot(2 * x, y, marker='*')
```

```python
fig, ax = plt.subplots()

x = np.array([-3, -2, -1, 0.1, 1, 2, 3])
y = np.array([9, 4, 1, 0.1, 1, 4, 9])

ax.plot(x, y, color='red', linestyle='--', marker='o', alpha=0.2)
ax.plot(0.5*x, y, '*--g', linewidth=2.5)
```

```python
# axis limits
fig, axes = plt.subplots()

x = np.array([-3, -2, -1, 0.1, 1, 2, 3])
y = np.array([9, 4, 1, 0.1, 1, 4, 9])

axes.plot(x, y, color='red', linestyle='--', marker='o')
axes.set_xlim(0, 5)
axes.set_ylim(2, 10)
```

```python
# To mark our plots

fig, ax = plt.subplots()

x = np.array([-3, -2, -1, 0.1, 1, 2, 3])
y = np.array([9, 4, 1, 0.1, 1, 4, 9])

ax.plot(x, y, color='red', linestyle='-', marker='o', label='y = x^2')
ax.plot(0.5 * x, y, color='green', linestyle='-', marker='*', label='y = 0.5 * x^2')

ax.set_title('Header of our plot', fontsize=20)

ax.set_xlabel("x")
ax.set_ylabel("y")

ax.legend(loc='lower left')
```

```python
# Clear plot
plt.cla()
# clear figure
plt.clf()

fix, ax = plt.subplots()

x = np.array([-3, -2, -1, 0.1, 1, 2, 3])
y = np.array([9, 4, 1, 0.1, 1, 4, 9])

ax.plot(x,y, color='red', linestyle='--', marker='o', label='y = x^2')

ax.set_xticks(np.arange(-3, 3, 0.5))
ax.set_yticks(np.arange(1, 10, 3))

# reduce right border
ax.spines['right'].set_visible(False)

# create grid
ax.grid(color='blue', linestyle='-', linewidth=2)
# ax.grid() # to cancel grid


```

## Plots Spicies

```python
# scatter plot
fig, ax = plt.subplots()

N = 50

x = np.random.rand(N)
y = np.random.rand(N)

colors = np.random.rand(N)

area = (np.random.rand(N) * 30) ** 2

ax.scatter(x, y, c=colors, s=area, alpha=0.5, marker='*', cmap='viridis')


```

```python
# show 3d scatters
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x,y, s=area, c=colors, alpha=0.5, marker='*', cmap='viridis')
```

## Bars (column diagram)

```python
fig, axes = plt.subplots(1, 2)

axes[0].bar([1, 2, 3], [3, 4, 5], color='red', label='vertical_bar', alpha=0.8)

axes[1].barh([0.5, 1, 2.5], [0, 1, 2], label='horizontal_bar')

axes[0].legend()
axes[1].legend()

```

# Stacked bar chat

```python
fig = plt.figure(figsize=(14, 10))

rect1 = plt.bar(np.arange(5), np.arange(5) ** 2, width=0.5, color="#1fb430")

rect2 = plt.bar(np.arange(5), np.arange(5) * 2, width=0.5, color='#1f77b4')

plt.legend((rect1[0], rect2[0]), ('Support', 'Freedom'))

```

## BoxPlot

```python
user_1 = [10, 3, 15, 21, 17, 14]
user_2 = [5, 13, 10, 7, 9, 12]

data = [user_1, user_2]

fig = plt.figure(figsize=(8,6))
ax = fig.add_axes([0, 0, 1, 1])

bp = ax.boxplot(data)

plt.xticks([1,2], ['user_1', 'user_2'])
```

## Area Chart

```python
fig, ax = plt.subplots()

x = np.arange(0, 100, 1)
y = x * 2

ax.fill_between(x, y, color="skyblue", alpha=0.2)
ax.plot(x, y, color='Slateblue', alpha=0.6)
ax.grid()
```

## Pie Chart - Circle

```python
fig, ax = plt.subplots(figsize=(8,8))
labels = ['Jan', 'Feb', 'March', 'April', 'May', 'June']
user_1 = [10, 3, 15, 21, 17, 14]

ax.pie(user_1, labels=labels, explode=(0.07, 0, 0, 0, 0, 0), autopct='%1.1f%%', startangle=130, shadow=True)
```

## Time Series Plot (Working with Pandas)

```python
import pandas as pd

plt.figure(figsize=(8, 6))
ts = pd.Series(np.random.randn(100), index=pd.date_range('1/1/2020', periods=100))

ts = ts.cumsum()
ts.plot()
```

## 3D plot

```python
# 3D Surface

import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)

X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False)
```
