## Filter using mask

```python
df[df['Age'] > 18]

df[df['Age'].isin([5, 10, 15])]

df[(df['Age'] == 5) | (df['Age'] == 10)]

df[df['Age'].notna()]

df['Age'].isna().sum()

df.loc[df['Age'].isna(), ['Name', 'Age']]
```

## Sorting values

```python
df.sort_values('Age').head(10)
```

```python
df.sort_values(['Age', 'Name'], ascending=[False, True]).head(5)
```

## Merge and Concat DataFrames

```python
# CONCAT
df2 = df.copy(deep=True)
# concat by string
cdf1 = pd.concat([df, df2])
cdf1.shape

# concat by column
cdf2 = pd.concat([df, df2], axis=1)
cdf2.shape
```

```python
# merge
mdf = pd.DataFrame(index=df.index)
mdf['PassengerId'] = df['PassengerId']

# using apply with lambda
mdf['evenId'] = mdf['PassengerId'].apply(lambda x: x % 2 == 0)



```

```python
# MERGE DataFrames

pd.merge(df, mdf, how='inner').head(5)
```

## Analytics functions

```python
# count of elements for each column
df.count()

# for specific columns
df['Age'].count()

# Average value
df['Age'].mean()

# Median value
df['Age'].median()

# returns Several statistic values
df['Age'].describe()

```

```python
# groupby using
df.groupby('Sex')['Age'].mean()
df.groupby('Sex')['Survived'].count()
df.groupby('Sex')['Age'].describe()

# Several conditions
df.groupby(['Sex', 'Survived'])['Age'].agg(['mean', 'median'])
```

```python
# value_counts (count of elements of each group)

df.value_counts('Sex')
# equals
df.groupby('Sex')['Sex'].count()
```

```python
# Correlation func -> table - exclude string fields like Name

df.corr(numeric_only=True)
```

## Change DataFrames

```python
tmp_df = df.copy(deep=True)
tmp_df['Pclass'] = 1
```

```python
tmp_df['isAdult'] = tmp_df['Age'] >= 18
tmp_df.head(5)
```

```python
tmp_df['isAdult_isSurvived'] = tmp_df['isAdult']

# Specfic condition
tmp_df.loc[tmp_df['Survived'] == 0, 'isAdult_isSurvived'] = False

tmp_df[:5]
```

```python
# Rename columns
tmp_df.rename(
    columns={'isAdult_isSurvived': 'AdultSurvived'}
).head(1)
```

```python
# Using functions

tmp_df.rename(columns=str.lower).head(1)
```

```python
# Using apply with lambda

tmp_df['lower_case_name'] = tmp_df['Name'].apply(lambda x: x.lower())
tmp_df.head(3)
```

```python
# Fast method

tmp_df['lower_case_name_2'] = tmp_df['Name'].str.lower()
tmp_df.head(3)
```
